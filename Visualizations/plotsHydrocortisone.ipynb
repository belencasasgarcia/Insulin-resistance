```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from paper_behavior_functions import (seaborn_style, 
                                      group_colors, figpath,
                                      FIGURE_WIDTH, FIGURE_HEIGHT)

# Load some things from paper_behavior_functions
figpath = figpath()
seaborn_style()
pal = group_colors()

color1=[0, 0.25, 0.5]
color2=[0.584, 0.84, 0.96]

# Create an array with the colors we want to use
colors = [[0.92,0.68,0.0353],[0.8,0,0.2]] #Red, yellow
colors1 = [[0.8,0,0.2], [0.92,0.68,0.0353]]
colors2 = [color1, color2]


# Set custom color palette
customPalette = sns.set_palette(sns.color_palette(colors))


customPalette1 = sns.set_palette(sns.color_palette(colors1))

#customPalette2 = sns.set_palette(sns.color_palette(colors2))

simulationsHC = pd.read_csv (r'HC_simulations2.csv') 
dataHC = pd.read_csv (r'HC_data.csv') 

dataHC['time_days']=dataHC['time']/24

simulationsHC['time_days']=simulationsHC['time']/24

dataHC["I_mean_highHC"]=dataHC["I_mean_highHC"]*6.9444*0.001
dataHC["I_SD_highHC"]=dataHC["I_SD_highHC"]*6.9444*0.001
dataHC["I_mean_low_HC"]=dataHC["I_mean_low_HC"]*6.9444*0.001
dataHC["I_SD_low_HC"]=dataHC["I_SD_low_HC"]*6.9444*0.001

simulationsHC['Imeasured_min_11mM']=simulationsHC['Imeasured_min_11mM']*6.9444*0.001
simulationsHC['Imeasured_max_11mM']=simulationsHC['Imeasured_max_11mM']*6.9444*0.001
simulationsHC['Imeasured_min_5p5mM']=simulationsHC['Imeasured_min_5p5mM']*6.9444*0.001
simulationsHC['Imeasured_max_5p5mM']=simulationsHC['Imeasured_max_5p5mM']*6.9444*0.001
simulationsHC['Imeasured_min_11mM_lowHC']=simulationsHC['Imeasured_min_11mM_lowHC']*6.9444*0.001
simulationsHC['Imeasured_max_11mM_lowHC']=simulationsHC['Imeasured_max_11mM_lowHC']*6.9444*0.001
simulationsHC['Imeasured_min_5p5mM_lowHC']=simulationsHC['Imeasured_min_5p5mM_lowHC']*6.9444*0.001
simulationsHC['Imeasured_max_5p5mM_lowHC']=simulationsHC['Imeasured_max_5p5mM_lowHC']*6.9444*0.001
```

```python
# Glucose days 1-3

simulation_GTT = simulationsHC[simulationsHC['time']<=48]
plt.close('all')

fig = sns.FacetGrid(data=dataHC, hue="Glycemic_level", palette=customPalette1,
                    sharex=True, sharey=True,
                    height=FIGURE_HEIGHT, aspect=(FIGURE_WIDTH/5)/FIGURE_HEIGHT)

#fig.map(plt.errorbar, "time_total", "Glucose_mean", "Glucose_SEM", fmt='none')
fig.map(plt.errorbar, "time_days", "G_mean_highHC", "G_SD_highHC", fmt='none', elinewidth=1, capsize=0.5, capthick=0.3)
sns.lineplot(data=dataHC,
             x='time_days', y='G_mean_highHC', hue="Glycemic_level", marker='o', err_style='bars',
             color=customPalette1, linewidth=0, ci=95, ax=fig.ax, legend=None)# overlay the simulated


fig.ax.fill_between(data=simulation_GTT, x='time_days', y1='Gmeasured_min_11mM', y2='Gmeasured_max_11mM', alpha=0.45, 
                    facecolor=colors[1])

#fig.ax.fill_between(data=simulation_GTT, x='time_days', y1='Gmeasured_min_5p5mM', y2='Gmeasured_max_5p5mM', alpha=0.5, 
#                    facecolor=colors[0])


fig.set_axis_labels('Time (days)', 'Glucose (mM)')
fig.ax.set_xticks([0,1,2])
fig.ax.set_xticklabels(['1','2','3'])
fig.ax.set_xlim(-0.3,2.3)
fig.ax.set_ylim(3,16)
fig.despine(trim=True)
plt.tight_layout(w_pad=-0.1)
fig.savefig(os.path.join(figpath, "training_glucose_HC_day1to3.pdf"))
```
